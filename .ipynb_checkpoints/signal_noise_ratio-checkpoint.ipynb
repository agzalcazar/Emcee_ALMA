```python
import emcee
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colorbar import Colorbar
import astropy.io.fits as pyfits
import scipy.optimize as opt
import corner
import os
import utils as myUtils
import scipy.integrate as integrate
import matplotlib.gridspec as gridspec
from matplotlib.patches import Ellipse
```

```python
path = r'G:\Mi unidad\Universidad\Master_Astro\TFM\ALMA_images\\'
sources = os.listdir(path)[0:-1]
```

```python
def s_n(flux):
    max_signal = np.nanmax(flux)
    std = myUtils.std_image(flux)
    return max_signal/std
```

```python
source_SN = {}
for i in range(0,len(sources)):
    fits_file  = pyfits.open(path+sources[i])
    
    if sources[i]=='cont_clean_AL-75.1_plot_briggs.fits':
        image_data = fits_file[0].data
    else:
        image_data = fits_file[0].data[0][0]
        
    image_data.shape
    fig, ax = plt.subplots()
    plt1 = ax.imshow(myUtils.zoom(image_data,35))
    fig.colorbar(ax=ax, mappable=plt1, orientation='vertical')
    ax.set_title('Source: '+ str(sources[i])+ '\nS/N: '+ str(s_n(myUtils.zoom(image_data,35))))
    plt.show()
    source_SN[sources[i]]=s_n(myUtils.zoom(image_data,35))
    

    
```
