```python
I = flux
x = x_grid
y = y_grid
M0 = I.sum()
x0 = (x*I).sum()/M0
y0 = (y*I).sum()/M0
Mxx = (x*x*I).sum()/M0 - x0*x0
Myy = (y*y*I).sum()/M0 - y0*y0
Mxy = (x*y*I).sum()/M0 - x0*y0
D = 2*(Mxx*Myy-Mxy*Mxy)
a = Myy/D
c = Mxx/D
b = -Mxy/D


# Find the area of one pixel expressed in grids to find amplitude A
   
theta = np.degrees(0.5*np.arctan(2*c/(b-a))) 
if theta<0:  # Not HW1 but the largest axis corresponds to theta.
    theta+=180
```

```python
def Tflux(theta, x,y,profile,rms,points = 100):
    if profile== 'B':
        l = lambda y, x: myUtils.model_twoD_Gaussian(theta, x, y)
    if profile== 'A':
        l = lambda y, x: myUtils.model_Sersic(theta, x, y)
    integr, err = integrate.dblquad(l, -np.inf, np.inf, -np.inf, np.inf)

    noise = np.random.normal(loc=0.0, scale=rms, size=myUtils.model_twoD_Gaussian(parameters_fit_50, x, y).shape[0])

    if profile == 'B':
        l = lambda y, x: myUtils.model_twoD_Gaussian(theta, x, y) + np.random.normal(loc=0.0, scale=rms, size=1)
    if profile == 'A':
        l = lambda y, x: myUtils.model_Sersic(theta, x, y)  + noise
    integral_noise = []
    for i in range(0,points):
        integral_noise .append(integrate.dblquad(l, -np.inf, np.inf, -np.inf, np.inf)[0])

    return integr, integral_noise
```
